```python
fhand = open('mbox.txt')  # read an external file

list1 = [line.split("From ")[1].split(" ", 1)[0] for line in fhand if
         line.startswith("From ")]  # get all "from email" addresses into a list

email_count_dict = dict()  # create a empty dictionary

for each_email in list1:  # count the frequency of each email and save it in dictionary
    email_count = list1.count(each_email)
    email_count_dict[each_email] = email_count

print("Number of emails from each 'From address' is :")
for x, y in email_count_dict.items():  # print the histogram table
    print(x, ":", y)

max_count = 0
min_count = 99999
min_list = []
for y in email_count_dict.values():  # find the min and max value from the histogram
    if y > max_count:
        max_count = y
    if y < min_count:
        min_count = y

print("\nhighest number of emails:")
for x, y in email_count_dict.items():  # find email having highest fequency
    if y == max_count:
        print(x, ":", y)

print("\nlowest number of emails:")
for x, y in email_count_dict.items():  # find email having lowest fequency
    if y == min_count:
        print(x, ":", y)
```

```python
#find the sum, min, max, average and standard deviation of the X-DSPAM-Confidence value
#do not use any built in methods for computing the statistics.

fhand = open('mbox.txt')      #read the external file

email_dist = dict()                     #create an empty dictionary

for line in fhand:                      #read line by line to check if a line contains "from email address"
    if line.startswith("From "):        #and add the email address to the dictionary
        line.strip()
        email = line.split("From ")[1].split(" ", 1)[0]
        if email not in email_dist.keys():
            email_dist[email] = []
    elif line.startswith("X-DSPAM-Confidence:"):    #read line by line to check if a line contains confidence value
        value = line.split(":")[1].strip()          #and add the confidence value to the corresponding email address
        email_dist[email].append(value)

#for x, y in email_dist.items():         #print all email address along with their confidence values
#    print(x, ":", y)

for x, y in email_dist.items():
    sum: float = 0.0                    #sum of all confidence values of an email initialized to zero
    max: float = 0.0                    #maximum of confidence values of an email initialized to zero
    mean: float = 0.0                   #avarage of confidence values of an email initialized to zero
    min: float = 1.0                    #minimum of confidence values of an email initialized to zero
    numerator: float = 0.0              #numarator will be used to calculate variance and standard deviation
    for n in y:
        sum = sum + float(n)            #claculate min, max and avg of confidence level for each email
        if float(n) > max:
            max = float(n)
        if float(n) < min:
            min = float(n)
    mean = sum/len(y)

    for m in y:                         #calculate variance
        numerator = numerator + (float(m) - mean) ** 2
        variance = numerator / len(y)

    stnd_dev = variance**0.5            #calculate standard deviation

    print("statistics for ", x, " is: ", "\n sum = ", round(sum, 2), "\n max = ", round(max, 2),
          "\n min = ", round(min, 2), "\n avg = ", round(mean, 2),
          "\n standard deviation = ", round(stnd_dev, 2), "\n")
```

```python
fhand = open('mbox.txt')                                # read an external file

mbox_dict = {'Message-ID': [],
             'Mime-Version': [],
             'Content-Transfer-Encoding': [],
             'Date': [],
             'X-Authentication-Warning': [],
             'To': [],
             'From': [],
             'Subject': [],
             'X-Content-Type-Outer-Envelope': [],
             'X-Content-Type-Message-Body': [],
             'Content-Type': [],
             'X-DSPAM-Result': [],
             'X-DSPAM-Processed': [],
             'X-DSPAM-Confidence': [],
             'X-DSPAM-Probability': []}

flag = False
for each_line in fhand:

    if each_line.startswith("Message-ID"):
        msg_id = each_line.split(" ", 1)[1].strip()
        mbox_dict["Message-ID"].append(msg_id)

    elif each_line.startswith("Mime-Version:"):
        mime_vers = each_line.split(":")[1].strip()
        mbox_dict["Mime-Version"].append(mime_vers)

    elif each_line.startswith("Content-Transfer-Encoding:"):
        cont_trans_code = each_line.split(":")[1].strip()
        mbox_dict["Content-Transfer-Encoding"].append(cont_trans_code)

    elif each_line.startswith("Date "):
        flag = not flag
        if flag:
            date = (each_line.split(",")[1]).strip()[0 : 11]
            mbox_dict["Date"].append(date)

    elif each_line.startswith("X-Authentication-Warning:"):
        auth_warn = each_line.split(":",1)[1].strip()
        mbox_dict["X-Authentication-Warning"].append(auth_warn)

    elif each_line.startswith("To:"):
        to = each_line.split(":")[1].strip()
        mbox_dict["To"].append(to)

    elif each_line.startswith("From:"):
        sender = each_line.split(":")[1].strip()
        mbox_dict["From"].append(sender)

    elif each_line.startswith("Subject:"):
        sub = each_line.split(":",1)[1].strip()
        mbox_dict["Subject"].append(sub)

    elif each_line.startswith("X-Content-Type-Outer-Envelope:"):
        content_envlop = each_line.split(":")[1].strip()
        mbox_dict["X-Content-Type-Outer-Envelope"].append(content_envlop)

    elif each_line.startswith("X-Content-Type-Message-Body:"):
        content_body = each_line.split(":")[1].strip()
        mbox_dict["X-Content-Type-Message-Body"].append(content_body)

    elif each_line.startswith("Content-Type:"):
        content_type = each_line.split(":")[1].strip()
        mbox_dict["Content-Type"].append(content_type)

    elif each_line.startswith("X-DSPAM-Result:"):
        spam_result = each_line.split(":")[1].strip()
        mbox_dict["X-DSPAM-Result"].append(spam_result)

    elif each_line.startswith("X-DSPAM-Processed:"):
        spam_process = each_line.split(":",1)[1].strip()
        mbox_dict["X-DSPAM-Processed"].append(spam_process)

    elif each_line.startswith("X-DSPAM-Confidence:"):
        confidence = each_line.split(":")[1].strip()
        mbox_dict["X-DSPAM-Confidence"].append(confidence)

    elif each_line.startswith("X-DSPAM-Probability:"):
        prob = each_line.split(":")[1].strip()
        mbox_dict["X-DSPAM-Probability"].append(prob)

for each_key,value in mbox_dict.items():
    print(each_key,len(value))
```
