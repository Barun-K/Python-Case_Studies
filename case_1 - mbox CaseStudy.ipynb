1).	Write a program to build a histogram using a dictionary to count how many messages have come from each email address, and print the dictionary.
---------------------------------

```python
fhand = open('mbox.txt')     # read an external file

c = 1                        # set counter for each email count initialization
email_histogram = dict()     # create an empty dictionary

for line in fhand:
    if line.startswith('From '):     
        line1 = line.split(' ')[1].split(' ')[0]        # read each emailID  
        if line1 not in email_histogram.keys():   
            email_histogram[line1] = c
        else:
            email_histogram[line1] = email_histogram[line1] + 1  # count the frequency of each emailID and save in dictionary

            
for x,y in email_histogram.items():
    print(x,':',y)                       # print histogram table ( count for each emailID )
```

1b).	For Every “From address” find the sum, min max, average and standard deviation of the X-DSPAM-Confidence value -  do not use any built in methods for computing the statistics. You are allowed to use map and reduce methods for computing the statistics.
-----------------------------

```python
from functools import reduce  # import reduce function for stats calculation

fhand = open('mbox.txt')     # read an external file

email_stats = dict()     # create an empty dictionary

for line in fhand:
    if line.startswith('From '):     
        emailID = line.split(' ')[1].strip()        # read each emailID  
        if emailID not in email_stats.keys():   
            email_stats[emailID] = []               # store each emailID in dictionary
    elif line.startswith('X-DSPAM-Confidence'):
        conval = float(line.split(':')[1].strip())
        email_stats[emailID].append(conval)         # append X-DSPAM-Confidence value to corresponding emailID
        
for x,y in email_stats.items():
    sum_cv = reduce(lambda a,b: a+b,y)              # calculate sum for X-DSPAM-Confidence value for corresponding emailID
    min_cv = reduce(lambda a,b: a if a<b else b,y)  # calculate min for X-DSPAM-Confidence value for corresponding emailID
    max_cv = reduce(lambda a,b: a if a>b else b,y)  # calculate max for X-DSPAM-Confidence value for corresponding emailID
    average_cv = sum_cv/len(y)                      # calculate average for X-DSPAM-Confidence value for corresponding emailID
    
    numerator = 0.0
    for m in y:                # calculate variance
        numerator = numerator + ((m - average_cv) ** 2)
    variance = numerator / len(y)

    sd_cv = variance**0.5            # calculate sd for X-DSPAM-Confidence value for corresponding emailID
    

    print("Statistics for X-DSPAM-Confidence value for [",x, "] is: ", 
          "\n sum = ", round(sum_cv, 2), "\n max = ", round(max_cv, 2),
          "\n min = ", round(min_cv, 2), "\n avg = ", round(average_cv, 2),
          "\n standard deviation = ", round(sd_cv, 2), "\n")
```

```python
fhand = open('mbox.txt')     # read an external file

c = 1                        # set counter for each email count initialization
email_dcount = dict()     # create an empty dictionary

for line in fhand:
    if line.startswith('From '):     
        domain = line.split('@')[1].split(' ')[0]         # collect each domain  
        if domain not in email_dcount.keys():   
            email_dcount[domain] = c
        else:
            email_dcount[domain] = email_dcount[domain] + 1  # count the frequency of each domain and save in dictionary

            
for x,y in email_dcount.items():
    print(x,':',y)                       # print histogram table ( count for each domain )
```

```python
# 3).	Write a program to convert mbox.txt into a dictionary with the following keys, 
#       Every individual value should appear as a list :
#       Message-ID, Mime-Version, Content-Transfer-Encoding,Date,X-Authentication-Warning,To,From,Subject,
#       X-Content-Type-Outer-Envelope,X-Content-Type-Message-Body,Content-Type,X-DSPAM-Result,
#       X-DSPAM-Processed,X-DSPAM-Confidence,X-DSPAM-Probability



fhand = open('mbox.txt')                                # read an external file

mbox_dict = {'Message-ID': [],
             'Mime-Version': [],
             'Content-Transfer-Encoding': [],
             'Date': [],
             'X-Authentication-Warning': [],
             'To': [],
             'From': [],
             'Subject': [],
             'X-Content-Type-Outer-Envelope': [],
             'X-Content-Type-Message-Body': [],
             'Content-Type': [],
             'X-DSPAM-Result': [],
             'X-DSPAM-Processed': [],
             'X-DSPAM-Confidence': [],
             'X-DSPAM-Probability': []}

flag = False
for each_line in fhand:

    if each_line.startswith("Message-ID"):
        msg_id = each_line.split(" ", 1)[1].strip()
        mbox_dict["Message-ID"].append(msg_id)

    elif each_line.startswith("Mime-Version:"):
        mime_vers = each_line.split(":")[1].strip()
        mbox_dict["Mime-Version"].append(mime_vers)

    elif each_line.startswith("Content-Transfer-Encoding:"):
        cont_trans_code = each_line.split(":")[1].strip()
        mbox_dict["Content-Transfer-Encoding"].append(cont_trans_code)

    elif each_line.startswith("Date "):
        flag = not flag
        if flag:
            date = (each_line.split(",")[1]).strip()[0 : 11]
            mbox_dict["Date"].append(date)

    elif each_line.startswith("X-Authentication-Warning:"):
        auth_warn = each_line.split(":",1)[1].strip()
        mbox_dict["X-Authentication-Warning"].append(auth_warn)

    elif each_line.startswith("To:"):
        to = each_line.split(":")[1].strip()
        mbox_dict["To"].append(to)

    elif each_line.startswith("From:"):
        sender = each_line.split(":")[1].strip()
        mbox_dict["From"].append(sender)

    elif each_line.startswith("Subject:"):
        sub = each_line.split(":",1)[1].strip()
        mbox_dict["Subject"].append(sub)

    elif each_line.startswith("X-Content-Type-Outer-Envelope:"):
        content_envlop = each_line.split(":")[1].strip()
        mbox_dict["X-Content-Type-Outer-Envelope"].append(content_envlop)

    elif each_line.startswith("X-Content-Type-Message-Body:"):
        content_body = each_line.split(":")[1].strip()
        mbox_dict["X-Content-Type-Message-Body"].append(content_body)

    elif each_line.startswith("Content-Type:"):
        content_type = each_line.split(":")[1].strip()
        mbox_dict["Content-Type"].append(content_type)

    elif each_line.startswith("X-DSPAM-Result:"):
        spam_result = each_line.split(":")[1].strip()
        mbox_dict["X-DSPAM-Result"].append(spam_result)

    elif each_line.startswith("X-DSPAM-Processed:"):
        spam_process = each_line.split(":",1)[1].strip()
        mbox_dict["X-DSPAM-Processed"].append(spam_process)

    elif each_line.startswith("X-DSPAM-Confidence:"):
        confidence = each_line.split(":")[1].strip()
        mbox_dict["X-DSPAM-Confidence"].append(confidence)

    elif each_line.startswith("X-DSPAM-Probability:"):
        prob = each_line.split(":")[1].strip()
        mbox_dict["X-DSPAM-Probability"].append(prob)

for each_key,value in mbox_dict.items():
    print(each_key,",[Length] : ",len(value),'\n[Values] :',value,"\n-----\n")
```

```python
#   4a).How do you sort a dictionary based on: Keys
#-------------------------------------------------------

numbers1 = {'one': 1, 'two': 2, 'three': 3, 'four': 4,'five': 5}

print('4a).How do you sort a dictionary based on: Keys')

print(sorted(numbers1))                            #simpleset way to sort a dictionary based on ascending order of keys 
print(sorted(numbers1.keys()))                     #alternatively we can specify to sort the dictionary using keys         
print(sorted(numbers1.keys(),reverse = True))      #sort the dictionary based on key's desending value
print(sorted(numbers1, key=numbers1.__getitem__))   #custom way to sort keys by comparing values of all the keys

#sorting using string function 
numbers2 = {'one': 1, 'two': 2, 'three': 3, 'four': 4,'Fourth': 4}

print('\n Sorting using string function ')
print(sorted(numbers2))                            #"Fourth" comes before "four" because of Capital initials
print(sorted(numbers2, key = str.lower))          #convert all capital letters to small while compairing:
                                                  #now "Fourth" comes after "four"



#    4b).How do you sort a dictionary based on: Values
#-------------------------------------------------------

numbers = {'one': 1, 'two': 2, 'three': 3, 'four': 4,'five': 5}

print('\n4b).How do you sort a dictionary based on: Values')
print(sorted(numbers.values()))                  #sort the dictionary based on list of values ascending 
print(sorted(numbers.values(),reverse = True))   #sort the dictionary based on reversed order of values i.e descending
print([(key,value) for key ,value in sorted(numbers.items())])  ##custom way to sort values by comparing of all the keys


```
