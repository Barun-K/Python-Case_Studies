21).	Find: a).	Mean and std deviation of the income
--------------

```python
oj.income.agg(['mean','std'])
```

b.	For income greater than or equal to 10.5, find the mean income
----------

```python
oj[oj.income>=10.5].agg({'income' : np.mean})
```

c.	For each brand having price >=2.5 find the mean, median, sd of the normalised value of income
----------

```python
oj['normInc'] = (oj.income - oj.income.mean())/oj.income.std()
oj[['income','normInc']][oj.income>=10.5].agg({'normInc' : [np.mean,np.median,np.std]})
```
